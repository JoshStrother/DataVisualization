```python
import pandas as pa
import matplotlib.pyplot as plt
import seaborn as sns

df = pa.read_csv('https://raw.githubusercontent.com/nurfnick/Data_Viz/main/Activity_Dataset_V1.csv') #workout data
df1 = pa.read_csv('https://raw.githubusercontent.com/nurfnick/Data_Viz/main/AB_NYC_2019.csv') #air bnb data
df.head()
```

First we imported our libraries and data.  I used the workout data for this first one because I really wanted to see the correlations.

```python
ax = sns.heatmap(df.corr(),vmin = -1)
ax.set(title = 'Heatmap of Workout Stats')

plt.show()
```

So I was hoping to see more here but really the correlations are between the totals, averages, and max/min.  I think more of what I was wanting was a scatterplot maybe but we have already done that with this in a previous lesson.

```python
df1.head() #air BNB data
```

For this next one I will use the Air BNB data.

```python
ax = sns.scatterplot(data = df1, x = 'price', y = 'availability_365', hue = 'neighbourhood_group', style= 'room_type')
ax.set(title = "Price vs Availability",
       xlabel = "Price",
       ylabel = "Days Available Per Year")

plt.show()
```

For this scatter plot I did the price by availability and then the loaction and room type.  This gives some obvious information such as the entire home or apt are more expensive.  But some information that I didn't know would be that the Bronx, Queens, and Staten Island are mostly under 2k.  There are a few outliers within Bronx and Staten island. There doesn't seem to be a direct correlation between the price and how often its available, just the price and how many listings there are.
