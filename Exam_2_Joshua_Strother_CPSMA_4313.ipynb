Now we need to remove the unnamed column.

```python
del df['Unnamed: 0']

df.head()
```

Using regular expression we can remove (%) and replace it with nothing.  I need to use the \ to escape the parenthesis because they are special characters.

```python
df = df.rename(columns=lambda x: re.sub('\(%\)','',x))
df.columns
```

For section d. we are converting activity_day to datetime

```python
df.activity_day = pa.to_datetime(df.activity_day, format='%Y-%m-%d')
```

For section e. we need to replace the NaN values, they should be 0 since it was exercises where there wasnt necessarily steps involved. So we will fill the values and change the type to int.

```python
df.total_steps = df.total_steps.fillna(value = 0)
df.total_steps = df.total_steps.astype(int)
df.total_steps
```

for section f we need to convert the data to a float and make sure that we keep the data integrity.  Im going to use a temp holder to get all non NaN values, then i will apply a conversion then set the main column equal to that temp.

```python
temp = df[df['avg_pace'].notnull()]
temp.avg_pace
```

```python
df['avg_pace'] = temp['avg_pace'].str.split(':').apply(lambda x: int(x[0]) + int(x[1])/60)
df.avg_pace
```

Now we want the mean, media, count, and std of calories when grouped by workout.

```python
df.groupby('workout_type').calories.agg(['mean','median','count', 'std'])
```

For part h we are adding a column that indicates if the aerbic percent was over 30.

```python
df['aer>30'] = df.aerobic > 30
df
```

Now we want to convert the date time to week days, gather the day, workout, and cadence then get the max value.

```python
df['day'] = df['activity_day'].dt.day_name()
cad = df[['day','workout_type','max_cadence']]
cad.max()
```
