First we need to get pandas and our data, so I will do that as one.

```python
import pandas as pa

df = pa.read_csv('https://raw.githubusercontent.com/nurfnick/Data_Viz/main/AB_NYC_2019.csv')

df.head()
```

Now as discussed in class, we know that the last_review column and the reviews_per_month column have NaN or NULL values.

```python
df.isnull().sum(axis = 0)
```

```python
df.loc[(df['number_of_reviews'] == 0)]
```

So when we counted the # of times that the two columns showed up NaN they both produced 10052 lines.  In our above statement we tested how many time that the number of reviews was 0 and it also came up to 10052 times.  This makes it safe to say that the cause of the NaN was due to never being reviewed before and in that case we can just change all of the Nan in the last review and reviews per month columns to 0.

```python
df.reviews_per_month = df.reviews_per_month.fillna(0)
df.last_review = df.last_review.fillna(0)

df
```

Now that I have changed all of the NaN to 0 I'm going to double check that the amount of NaN has actually changed and that it was done correctly.

```python
df.isnull().sum(axis = 0)
```

Cool, looks like it did indeed remove all NaN from the last_review and the reviews_per_month columns.
