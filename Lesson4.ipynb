Next for the actual assignment we want to find the average of a trip based on where it starts, so I want to see the # of trips and the total # of time taken from that starting point.

Edit:  I kept getting an error trying to display both the start_station_id and the start_station_name so I will run 2 different sets of code, one on just the id and one on just the name...

```python
%%bigquery --project data-viz-338716
SELECT start_station_id, COUNT(start_station_id) as num_of_trips, SUM(duration_minutes) as total_time
FROM `bigquery-public-data.austin_bikeshare.bikeshare_trips`
GROUP BY start_station_id
```

```python
%%bigquery --project data-viz-338716
SELECT start_station_name, COUNT(start_station_name) as num_of_trips, SUM(duration_minutes) as total_time
FROM `bigquery-public-data.austin_bikeshare.bikeshare_trips`
GROUP BY start_station_name
```

Just as I suspected.... it appears that either A. some of the station names have more than one name attached to it OR B.  There are more than one station on a street/intersection, this would require me to go through the actual full set of data in order to either confirm or deny it seems.

Edit: after trying to calculate the average trip time using the ID, one ID comes back as NaN and I can not for the life of me figure out how to get around that, so I will continue by using the station names.

```python
%%bigquery --project data-viz-338716
SELECT start_station_name, SUM(duration_minutes)/COUNT(start_station_name) as avg_trip_time
FROM `bigquery-public-data.austin_bikeshare.bikeshare_trips`
GROUP BY start_station_name
```
