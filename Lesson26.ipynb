First we need to import pandas to read the data, the data, and seaborn to visualize it.

```python
import seaborn as sns
import pandas as pa

df = pa.read_csv('https://raw.githubusercontent.com/nurfnick/Data_Viz/main/Activity_Dataset_V1.csv')
```

For the first boxplot we need the calories.  I want to try to make it match the same as pandas will do, so we will try to change the color of the lines to make it exact.

```python
rgb= [0,95,250]
rgbscaled = []
for color in rgb:
  rgbscaled.append(color/256)
rgbscaled

ax = sns.boxplot(y=df['calories'], color = 'white', linewidth=0.75,  
                 medianprops=dict(color="green", alpha=1), 
                 whiskerprops=dict(color=rgbscaled, alpha=1), 
                 capprops=dict(color=rgbscaled, alpha=1), 
                 boxprops=dict(edgecolor=rgbscaled))
```

At first to attempt this I was using some get_line()[a::b] method but I did not understand it, and then I found this direct assign method which is much easier to understand but a little more to type.  This graph would look more exact if I could figure out how to horizontally shrink the box.

```python
ax = sns.boxplot(y=df['calories'], x=df['workout_type'], color = 'white', linewidth=0.75,  
                 medianprops=dict(color="green", alpha=1), 
                 whiskerprops=dict(color=rgbscaled, alpha=1), 
                 capprops=dict(color=rgbscaled, alpha=1), 
                 boxprops=dict(edgecolor=rgbscaled))
```

This is decent, but it is not in the same order as in pandas.  I tried a group by but couldn't figure that out.  I just found out there is an order parameter I will try.

```python
ax = sns.boxplot(y=df['calories'], x=df['workout_type'], color = 'white', linewidth=0.75, 
                 order = ["Cricket", "Freestyle", "Indoor Cycling", "Open Water", "Outdoor Cycling", "Outdoor Running", "Pool Swimming", "Trail Run", "Treadmill", "Trekking", "Walking"],   
                 medianprops=dict(color="green", alpha=1), 
                 whiskerprops=dict(color=rgbscaled, alpha=1), 
                 capprops=dict(color=rgbscaled, alpha=1), 
                 boxprops=dict(edgecolor=rgbscaled))
```

Okay so it does appear that the order changed, now we need to rotate the names.  We can do this with the rotation parameter using set_xlabels.

```python
ax = sns.boxplot(y=df['calories'], x=df['workout_type'], color = 'white', linewidth=0.75,
                 order = ["Cricket", "Freestyle", "Indoor Cycling", "Open Water", "Outdoor Cycling", "Outdoor Running", "Pool Swimming", "Trail Run", "Treadmill", "Trekking", "Walking"],
                 medianprops=dict(color="green", alpha=1), 
                 whiskerprops=dict(color=rgbscaled, alpha=1), 
                 capprops=dict(color=rgbscaled, alpha=1), 
                 boxprops=dict(edgecolor=rgbscaled))

ax.set_xticklabels(ax.get_xticklabels(), rotation=45)
```

For the rotation of the names I used source 2.  Now we want the box plot horizontal..we can easily do this by swapping the x and y values. 

```python
ax = sns.boxplot(x=df['calories'], y=df['workout_type'], color = 'white', linewidth=0.75,
                 order = ["Cricket", "Freestyle", "Indoor Cycling", "Open Water", "Outdoor Cycling", "Outdoor Running", "Pool Swimming", "Trail Run", "Treadmill", "Trekking", "Walking"],
                 medianprops=dict(color="green", alpha=1), 
                 whiskerprops=dict(color=rgbscaled, alpha=1), 
                 capprops=dict(color=rgbscaled, alpha=1), 
                 boxprops=dict(edgecolor=rgbscaled))

ax.set_xticklabels(ax.get_xticklabels(), rotation=45)
```

We also want a jitter so we can add that with sns.stripplot.

```python
ax = sns.boxplot(x=df['calories'], y=df['workout_type'], color = 'white', linewidth=0.75,
                 order = ["Cricket", "Freestyle", "Indoor Cycling", "Open Water", "Outdoor Cycling", "Outdoor Running", "Pool Swimming", "Trail Run", "Treadmill", "Trekking", "Walking"],
                 medianprops=dict(color="green", alpha=1), 
                 whiskerprops=dict(color=rgbscaled, alpha=1), 
                 capprops=dict(color=rgbscaled, alpha=1), 
                 boxprops=dict(edgecolor=rgbscaled))

sns.stripplot(data = df, y = 'workout_type', x = 'calories' , color = 'grey')

ax.set_xticklabels(ax.get_xticklabels(), rotation=45)
```
