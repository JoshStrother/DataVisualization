```python
import matplotlib
import pandas as pa
```

Now we can import our data with pandas.

```python
df = pa.read_csv('https://raw.githubusercontent.com/nurfnick/Data_Viz/main/AB_NYC_2019.csv')
```

And for the first part we want to plot the max price based on the neighbourhood.

```python
ax = df.groupby('neighbourhood_group').price.agg('max').sort_values(ascending = True).plot(kind = 'barh', color = 'orange', edgecolor = 'blue')

for container in ax.containers:
    ax.bar_label(container)
```

Now we want to group based on the neighbourhood and the room type then the average price in that location.

```python
ax = df.groupby(['neighbourhood_group', 'room_type']).price.agg('mean').plot(kind = 'barh')

for container in ax.containers:
    ax.bar_label(container)
```

I wanted this to show neighbour hood on the axis then within that have the 3 different groups on the graph in a different color... so im going to try to get only the 3 columns we will use and then graph it.

```python
dftest = df[['neighbourhood_group', 'room_type', 'price']]
ax = dftest.groupby('neighbourhood_group').agg('mean').plot(kind = 'barh')

for container in ax.containers:
    ax.bar_label(container)
```

Okay that just didn't show the room type at all...I'll have to come back to this.

```python
df_pivot = pa.pivot_table(
    df, 
    values="price",
    index="neighbourhood_group",
    columns="room_type", 
    aggfunc=max
)

df_pivot.plot.bar(title = 'Grouped by Borough')
```
