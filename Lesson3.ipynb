**Testing BigQuery**

Here I am pulling the total amount of rows because it is the first part of the assignment.  But, I would also like to make sure that I can actually get BigQuery to work since it is my first time ever using it.

```python
%%bigquery --project data-viz-338716
SELECT 
  COUNT(*) as total_rows
FROM `bigquery-public-data.austin_bikeshare.bikeshare_trips`
```

Since I am unsure of what data I actually have I will use SELECT* to see every column in the table and limit it at 5 rows so I don't waste resources.

```python
%%bigquery --project data-viz-338716
SELECT *
FROM `bigquery-public-data.austin_bikeshare.bikeshare_trips`
LIMIT 5
```

The next part of the assignment asks that we find the longest trip so I will use the MAX function.

```python
%%bigquery --project data-viz-338716
SELECT MAX(duration_minutes) as longest_time_in_minutes
FROM `bigquery-public-data.austin_bikeshare.bikeshare_trips`
```

That seemed like a VERY long trip so I was not sure if I had gathered the right information.  I then decided to list the top 5 longest trips just to see if that was accurate and remembered to gather the trip_id this time.

```python
%%bigquery --project data-viz-338716
SELECT trip_id, duration_minutes
FROM `bigquery-public-data.austin_bikeshare.bikeshare_trips`
ORDER BY duration_minutes DESC
LIMIT 5
```

Okay so it seems to check out as long as I either did the right thing twice or the wrong thing twice.  Now we need to compute the average so instead of using the MAX function I will use the AVG function.

```python
%%bigquery --project data-viz-338716
SELECT AVG(duration_minutes) as average_trip_time
FROM `bigquery-public-data.austin_bikeshare.bikeshare_trips`
```

I was actually surprised to see that the average time was ~31 minutes considering the longest trips were so high.  I am now going to see the top 100 longest and shortest trips to see whats going on with that.

```python
%%bigquery --project data-viz-338716
SELECT trip_id, duration_minutes
FROM `bigquery-public-data.austin_bikeshare.bikeshare_trips`
ORDER BY duration_minutes DESC
LIMIT 100
```
