```python
import pandas as pa
import matplotlib.pyplot as plt
import seaborn as sns

df = pa.read_csv('https://raw.githubusercontent.com/nurfnick/Data_Viz/main/Activity_Dataset_V1.csv')
df.head()
```

The standard error is the how accurate the mean of any given sample is, while the confidence interval is how confident the data is to fall between 2 values

```python
dfgrouped = df.groupby('workout_type').agg(['mean','std', 'count'])

dfgrouped.calories.plot.bar(y = 'mean',yerr = 'std', legend = False, xlabel = 'Workout Type', ylabel='Calories Burned',   color = ['pink', 'red', 'orange', 'gold',  'yellow', 'green', 'cyan', 'blue', 'indigo', 'violet', 'brown'], title = "Error of Standard Deviation")
```
